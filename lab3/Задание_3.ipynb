```python
Q1 = df[' loan_term'].quantile(0.25)
Q3 = df[' loan_term'].quantile(0.75)
IQR = Q3-Q1
print(Q1)
print(Q3)
print(IQR)
```

```python
lower_fence = Q1 - 5.5 * IQR
upper_fence = Q3 + 5.5 * IQR
outliers = df[(df[' loan_term'] < lower_fence) | (df[' loan_term'] > upper_fence)]
#outliers = df[(df < lower_fence) | (df > upper_fence)]
print("Выбросы:")
print(outliers)
```

Как я понял выбросы, которые я искусственно ввёл считаются псевдовыбросами и в контексте данных они не являются выбросами. Датасет прекрасен и чист.

```python

data = df[' loan_term']

mean = data.mean()
std = data.std()

lower_fence = mean - 3 * std
upper_fence = mean + 3 * std

outliers = data[(data < lower_fence) | (data > upper_fence)]

print("Выбросы:")
print(outliers)
```

Методом среднего получилось выявить псевдовыбросы.  Что ранее считалось выбросами, стало считаться аномалиями, потому что среднее значение и стандартное отклонение изменились после изменения данных.

###Пример реализации

```python
import pandas as pd
import numpy as np


def corr_pearson(column_1, column_2):
  cov = np.cov(column_1, column_2)
  return cov/(column_1.std() * column_2.std())
```

```python
data = {
    'Рост': [170, 155, 180, 162, 175],
    'Вес': [65, 50, 75, 55, 70]
}
df = pd.DataFrame(data)
df
```

```python
corr_pearson(df["Рост"], df["Вес"])
```
