```python
df.columns = df.columns.str.strip()
```

```python
from linearmodels import PanelOLS
import statsmodels.api as sm

# Для начала преобразуем данные в формат панели
# Необходимо, чтобы данные имели MultiIndex с уровнями компании и года
df.set_index(['Company', 'Year'], inplace=True)

# Определяем зависимую и независимые переменные
dependent_var = df['Revenue']
independent_vars = df[['Market Cap(in B USD)', 'Gross Profit', 'Net Income', 'EBITDA',
                         'Share Holder Equity', 'Debt/Equity Ratio', 'ROE', 'Net Profit Margin']]

# Добавляем константу для моделей
independent_vars = sm.add_constant(independent_vars)
```

```python
#модель с фиксированными эффектами (FE)
fe_model = PanelOLS(dependent_var, independent_vars, entity_effects=True, time_effects=True)
fe_results = fe_model.fit()

fe_results.summary
```

```python
#модель со случайными эффектами (RE)
from linearmodels.panel import RandomEffects

re_model = RandomEffects(dependent_var, independent_vars)
re_results = re_model.fit()

re_results.summary
```

```python
#объединённую модель (Pooling Model)
pooled_model = PanelOLS(dependent_var, independent_vars)
pooled_results = pooled_model.fit()

pooled_results.summary
```

1. Общие характеристики модели:

R-squared (Between): 0.9239 — это коэффициент детерминации между объектами (компаниями), показывающий, насколько хорошо модель объясняет различия между объектами.
R-squared (Within): 0.9594 — это коэффициент, показывающий, насколько хорошо модель объясняет различия в динамике во времени для одного объекта.
R-squared (Overall): 0.9383 — общий показатель, показывающий, как модель объясняет общую вариацию данных.
No. Observations (Число наблюдений): 161 — общее количество наблюдений в данных.
Entities (Количество объектов): 12 — количество компаний.
Time periods: 15 — количество временных периодов (лет).
F-statistic: 288.97 — значение F-критерия, который показывает, насколько значимы все регрессоры (независимые переменные) совместно. P-value равен 0.0000, что свидетельствует о высокой значимости модели.
2. Параметры модели (Parameter Estimates):
Это ключевая часть отчета, которая показывает, как независимые переменные влияют на зависимую переменную.

Столбцы:
Parameter (Параметр): Оценки коэффициентов для каждой независимой переменной и константы.
Std. Err. (Стандартная ошибка): Показывает неопределенность в оценке коэффициента. Чем меньше стандартная ошибка, тем точнее оценка.
T-stat: T-статистика для проверки гипотезы, что коэффициент равен нулю. Чем больше t-статистика по модулю, тем более значима переменная.
P-value: P-значение для t-статистики. Если P-value меньше 0.05, то коэффициент значим на уровне 5%.
Lower CI, Upper CI (Доверительный интервал): Нижняя и верхняя границы 95% доверительного интервала для коэффициента.
Разбор параметров:
const (Константа): 12570.0 — базовое значение дохода компании, когда все независимые переменные равны нулю.
T-stat = 3.7286, P-value = 0.0003: Значима на уровне 5%.
Market Cap(in B USD): -13.294 — при увеличении рыночной капитализации на 1 миллиард долларов, доход компании уменьшается на 13.294 единицы. Однако переменная незначима на уровне 5% (P-value = 0.1042).
Gross Profit: 2.1101 — валовая прибыль сильно и положительно влияет на доход. При увеличении валовой прибыли на единицу, доход увеличивается на 2.11. Переменная значима (P-value = 0.0000).
Net Income: -0.8811 — чистая прибыль оказывает отрицательное влияние на доход (на каждый доллар чистой прибыли доход уменьшается на 0.88), и она значима (P-value = 0.0393).
EBITDA: 1.2154 — увеличение EBITDA на единицу увеличивает доход на 1.2154. Переменная значима (P-value = 0.0009).
Share Holder Equity: -0.2196 — увеличение доли акционеров на единицу уменьшает доход на 0.2196. Переменная значима (P-value = 0.0001).
Debt/Equity Ratio: -2940.6 — соотношение долга к капиталу сильно отрицательно влияет на доход (уменьшает на 2940.6 единиц при увеличении отношения на единицу). Переменная значима (P-value = 0.0072).
ROE: 152.78 — рост рентабельности капитала на 1% увеличивает доход на 152.78. Переменная значима (P-value = 0.0065).
Net Profit Margin: -896.63 — увеличение чистой маржи снижает доход на 896.63. Переменная значима (P-value = 0.0000).
3. Выводы:
Модель объясняет значительную часть вариации доходов компании (93.83%).
Некоторые переменные, такие как валовая прибыль, EBITDA, ROE, и чистая маржа, оказывают сильное влияние на доход.
Рыночная капитализация незначима в этой модели (P-value > 0.05), но другие переменные, такие как валовая прибыль и чистая маржа, играют важную роль.

```python
# Перестроим модели, исключив переменную "Market Cap(in B USD)"
independent_vars_reduced = independent_vars.drop(columns=['Market Cap(in B USD)'])

# Модель с фиксированными эффектами (FE)
fe_model_reduced = PanelOLS(dependent_var, independent_vars_reduced, entity_effects=True)
fe_results_reduced = fe_model_reduced.fit()

# Модель со случайными эффектами (RE)
re_model_reduced = RandomEffects(dependent_var, independent_vars_reduced)
re_results_reduced = re_model_reduced.fit()

# Пула (Pooling Model)
pooled_model_reduced = PanelOLS(dependent_var, independent_vars_reduced)
pooled_results_reduced = pooled_model_reduced.fit()

# Вывод результатов
print(fe_results_reduced.summary)
print(re_results_reduced.summary)
print(pooled_results_reduced.summary)
```
